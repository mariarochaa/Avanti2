2. Faça a leitura de uma imagem de entrada 2D colorida qualquer e aplique o
filtro da média, mediana e gaussiano na imagem de entrada. Mostre os
resultados de cada filtro por meio da biblioteca matplotlib.

```python
import cv2
from matplotlib import pyplot as plt

image_path = '/content/D.jpg'
image = cv2.imread(image_path)


image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.show
```

```python
def filtro(image):
    mediana = cv2.blur(image, (3,3))
    media = cv2.medianBlur(image, 3)
    gaussiano = cv2.GaussianBlur(image, (3,3),0)

    return mediana, media, gaussiano

mediana, media, gaussiano = filtro(image_rgb)

plt.subplot(2, 2, 1)
plt.imshow(image_rgb)
plt.title('Imagem Original')

plt.subplot(2, 2, 2)
plt.imshow(media)
plt.title('Imagem Media')

plt.subplot(2, 2, 3)
plt.imshow(mediana)
plt.title('Imagem Mediana')

plt.subplot(2, 2, 4)
plt.imshow(gaussiano)
plt.title('Imagem Gaussiano')
```
